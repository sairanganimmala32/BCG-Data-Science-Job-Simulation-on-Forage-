```python
def plot_stacked_bars(dataframe,title_,size_=(18,10),rot_=0,legend_='upper right'):
    #plot staced bars with annotations
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )
    
    # Annote bars
    annotate_stacked_bars(ax,textsize=14)
    #Rename legend
    plt.legend(['Retension','churn'],loc=legend_)
    #labels
    plt.ylabel("company base (%)")
    plt.show()

def annotate_stacked_bars(ax,pad=0.99, colour="white",textsize=13):
    # Add value to annotatio bars
    # Iterate over the plotted rectangles/bars
    for p in  ax.patches:
        # calculate annotation
        value=str(round(p.get_height(),1))
        #if value is  0 do not  annotate
        if value=='0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+p.get_width()/2)*pad-0.05,(p.get_y()+p.get_height()/2*pad)),
             color=colour,
             size=textsize
        )
```

```python
def plot_distribution(dataframe,column,ax,bins_=50):
    #plot distribution of numerical varibale in a stacked histogram of retention and churn
    #creating a dataframe as churn and return columns
    temp=pd.DataFrame({"retention":dataframe[dataframe['churn']==0][column],
                       "churn":dataframe[dataframe["churn"]==1][column]})
    #plot the histogram
    temp[['retention','churn']].plot(kind='hist',bins=bins_,ax=ax,stacked='True')
    #x-axis label
    ax.set_xlabel(column)
    #change the x-axis to plain style
    ax.ticklabel_format(style="plain",axis='x')
```

# Feature Engineering

#### Difference between off-peak prices in  December and Preceding January


```python
# Group off-peak prices by companies and month
monthly_price_by_id = price_df.groupby(['id', 'price_date']).agg({'price_off_peak_var': 'mean', 'price_off_peak_fix': 'mean'}).reset_index()
    
# Get january and december prices
jan_prices = monthly_price_by_id.groupby('id').first().reset_index()
dec_prices = monthly_price_by_id.groupby('id').last().reset_index()

# Calculate the difference
diff = pd.merge(dec_prices.rename(columns={'price_off_peak_var': 'dec_1', 'price_off_peak_fix': 'dec_2'}), jan_prices.drop(columns='price_date'), on='id')
diff['offpeak_diff_dec_january_energy'] = diff['dec_1'] - diff['price_off_peak_var']
diff['offpeak_diff_dec_january_power'] = diff['dec_2'] - diff['price_off_peak_fix']
diff = diff[['id', 'offpeak_diff_dec_january_energy','offpeak_diff_dec_january_power']]
diff.head()
```

```python
client_df=client_df.merge(diff,on='id')
```

# Price flutuation

```python
price_df.head()
```

```python
price_df['price_date']=pd.to_datetime(price_df['price_date'],format='%Y-%m-%d')
```

```python
#grouping by the prices by companies and month
montly_price=price_df.groupby(['id','price_date']).agg({'net_price_off_peak':'mean','net_price_peak':'mean','net_price_mid_peak':'mean'}).reset_index()
montly_price.head()
```

```python
jan_prices=montly_price.groupby('id').first().reset_index()
dec_prices=montly_price.groupby('id').last().reset_index()
dec_prices=dec_prices.rename(columns={'net_price_off_peak':'dec_off_peak','net_price_peak':'dec_peak','net_price_mid_peak':'dec_mid_peak'})
dec_prices=dec_prices.drop('price_date',axis=1)
jan_prices=jan_prices.drop('price_date',axis=1)
diff=pd.merge(dec_prices,jan_prices,on='id')
diff['price_diff_off_peak']=diff['dec_off_peak']-diff['net_price_off_peak']
diff['price_diff_peak']=diff['dec_peak']-diff['net_price_peak']
diff['price_diff_mid_peak']=diff['dec_mid_peak']-diff['net_price_mid_peak']
diff=diff[['id','price_diff_off_peak','price_diff_peak','price_diff_mid_peak']]
diff.head()
```

```python
client_df=client_df.merge(diff,on='id')
client_df.head()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
